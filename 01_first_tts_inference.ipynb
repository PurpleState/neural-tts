### Using `gTTS` (Google Text-to-Speech) as an alternative

Since `TTS` does not support Python 3.12, we can use `gTTS` as an alternative. `gTTS` is a Python library and CLI tool to interface with Google Translate's text-to-speech API.

```python
pip install gTTS
```

```python
from gtts import gTTS
from IPython.display import Audio

text = "Hello, this is a test of the Google Text-to-Speech library."
tts = gTTS(text=text, lang='en')
tts.save("output.mp3")

Audio("output.mp3")
```

```python
from gtts.lang import tts_langs

# Get a dictionary of supported languages
supported_languages = tts_langs()

print("Supported languages for gTTS:")
for lang_code, lang_name in supported_languages.items():
    print(f"  {lang_code}: {lang_name}")
```
